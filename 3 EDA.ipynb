### Exploratory Data Analysis

```python
lagos.describe()
```

Let's check the distribution of some of the columns, our data contains outliers, so we would use the median

```python
plt.figure(figsize=[10, 7])
plt.title("Distribution of prices")
sns.histplot(lagos["Price"], kde=True, color="#000066")
plt.xlabel("Price (in millions)");
```

```python
import plotly.express as px
```

```python
px.histogram(data_frame=lagos, x="Price",marginal="box", 
             nbins=120, title="Distribution of Prices")
```

It seems the average price for sale of a house in Lagos is about 55 million naira

```python
px.histogram(data_frame=lagos, x="Price",marginal="box", nbins=120,
             color="Newly Built", title="Disribtion of Prices based on when Built")
```

The average price for a Newly Built house is 130 million naira, while the avergage price for one that isn't newly built is 18 million naira. 

This statistic shows newly built tends to be more expensive in the market

```python
px.histogram(data_frame=lagos, x="Price",marginal="box", nbins=120,
             color="City", title="Disribtion of Prices based on City")
```

As we can see in the Lagos Real Estate market, location does matter. 

Houses mostly in Ikoyi and Ikeja tend to be more expensive, since this where most elite live

```python
px.histogram(data_frame=lagos, x="Price",marginal="box", nbins=120,
             color="Type", title="Disribtion of Prices based on Property Type")
```

Built Houses are more expensive, followed by apartments and land, which makes sense

```python
plt.figure(figsize=(10,5))
plt.title("Distribution of Bedrooms")

sns.countplot(x=lagos["Bedrooms"]);
```

The properties that have 0 are mostly land properties and we can see that they are the most from our dataset, while houses with 5 bedrooms (mostly duplexes) come second.

```python
plt.figure(figsize=(10,5))
plt.title("Distribution of Bathrooms")

sns.countplot(x=lagos["Bathrooms"]);
```

```python
plt.figure(figsize=(10,5))
plt.title("Distribution of Toilets")

sns.countplot(x=lagos["Toilets"]);
```

Also there isn't any house with 4 toilets, this is strange because a normal home would have an extra toilet (guest toilet).

But as we compare the graphs of toilets and bathroom, we see the houses with 2 and 3 bathrooms have the same count for toilets.

```python
plt.figure(figsize=(10,5))
plt.title("Distribution of City")

sns.countplot(y=lagos["City"]);
```

Top 10 Most Populated Neighborhoods

```python
top_neigh = lagos[lagos["Neighborhood"] !="Other Lekki"]["Neighborhood"].value_counts().head(10)
```

```python
top_neigh
```

```python
plt.figure(figsize=[10,6])

plt.title("Top 10 most populated neighborhoods in Lagos")
sns.barplot(y=top_neigh.index, x=top_neigh)

plt.ylabel("Neighborhood")
plt.xlabel("Number of houses");
```
